Write a **recursive** program to print all elements in the list.

```python

def print_list(list, idx=0):
    if (idx == len(list)):
        return
    print(list[idx])
    print_list(list, idx + 1)
Cities = ["Delhi", "Mumbai", "Haryana", "Punjab"] #list

print(Cities)
```
