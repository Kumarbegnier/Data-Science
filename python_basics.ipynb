**Comparison**

Python uses these operators to compare values:


| Operator | Meaning                  |
| -------- | ------------------------ |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |


```python
a = 10
b = 3

print(a == b)   # False
print(a != b)   # True
print(a > b)    # False
print(a < b)    # True
print(a >= 10)  # True
print(b <= 15)  # False
```

**Comparing Strings**

Strings are compared **lexicographically** (dictionary order).

```python
print("apple" == "apple")   # True
print("apple" < "banana")  # True
print("Zoo" > "apple")     # False (ASCII-based)

# ‚ö†Ô∏è Capital letters come before lowercase in ASCII.
```

**Comparing Numbers and Variables**



```python
x = 5
y = 5.0

print(x == y)   # True (value is same)
print(x is y)   # False (different memory location)
```

**Deep vs Shallow Comparison**
üîπ Shallow Comparison (==)
üîπ Deep Comparison (Nested Objects)

**üîπ Shallow Comparison (==)**

Checks values at the top level.

```python
a = [[1, 2], [3, 4]]
b = [[1, 2], [3, 4]]

print(a == b)   # True
```

**üîπ Deep Comparison (Nested Objects)**

Use copy or deepcopy to understand behavior.

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

print(a == b)   # True
print(a is b)   # False
```

üëâ == ‚Üí content
üëâ is ‚Üí memory

Comparing Objects (Custom Class)
By default, objects are compared by **memory**.

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

s1 = Student(90)
s2 = Student(90)

print(s1 == s2)  # False
```

‚úîÔ∏è Override __eq__ for Value Comparison

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

s1 = Student(90)
s2 = Student(90)

print(s1 == s2)  # True

#üî• Very common interview question.
```

**Comparing Dictionaries**
Order doesn‚Äôt matter in dictionary comparison.

```python
d1 = {"a": 1, "b": 2}
d2 = {"b": 2, "a": 1}

print(d1 == d2)  # True
```

Compare Only Keys or Values

```python
print(d1.keys() == d2.keys())      # True
print(d1.values() == d2.values())  # True
```

**Floating-Point** Comparison (Important ‚ö†Ô∏è)

```python
print(0.1 + 0.2 == 0.3)  # False ‚ùå
# Never compare floats directly.
```

```python
# Correct Way
import math

print(math.isclose(0.1 + 0.2, 0.3))  # True
```

Using operator Module
Cleaner and faster comparisons.

```python
import operator

print(operator.eq(10, 10))  # True
print(operator.gt(5, 3))    # True
```

**Sorting vs Comparison**

Sorting uses comparison internally.

```python
nums = [5, 1, 8, 3]
nums.sort()

print(nums)
```

```python
#Custom Sort Comparison
students = [("Neeraj", 85), ("Amit", 90), ("Ravi", 80)]

students.sort(key=lambda x: x[1])
print(students)
```
