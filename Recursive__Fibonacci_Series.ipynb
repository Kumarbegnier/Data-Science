Example 2: Fibonacci_Series

```python
def Fibonacci_Series(n):
    if n <= 0:
        print("Incorrect input")
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return Fibonacci_Series(n-1)+Fibonacci_Series(n-2)
print(Fibonacci_Series(6))
```
