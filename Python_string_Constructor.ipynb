üîπ 5. Creating String Using str() **Constructor** (Python)


In Python, strings can be created using the str() constructor, which converts other data types into string format.


üî∏ Syntax

    str(object)


üî∏ Examples

    1Ô∏è‚É£ Creating a string from an integer


```
# This is formatted as code
```




````python
num = 100       #int
s = str(num)    # str() constructor = converts other data types int into string
print(s)        # print string
print(type(s))  # <class 'str'>
````

2Ô∏è‚É£ Creating a string from a float

````python
pi = 3.14         #float
s = str(pi)       # str() constructor = converts other data types Float into string
print(s)          # print string
print(type(s))    # <class 'str'>
````

3Ô∏è‚É£ Creating a string from a boolean

````python
flag = True    #boolean
s = str(flag)  #convert boolean into string
print(s)       #print string
print(type(s))    # <class 'str'>
````

4Ô∏è‚É£ Creating a string from a list

````python
lst = [1, 2, 3]
s = str(lst)
print(s)
````

5Ô∏è‚É£ Creating an empty string

````python
s = str()
print(s)
````
