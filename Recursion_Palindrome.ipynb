**Palindrome** Check using Recursion (Python)

A palindrome is a string that reads the same forward and backward.
Examples: madam, racecar, 121

*Recursive Logic (Interview Explanation)*

**Base Case:**

If the string length is **0 or 1**, it is a palindrome.

**Recursive Case:**

  1. Compare the **first** and **last** characters.

  2. If they are equal, recursively check the remaining substring.

  3. If not equal, return False.

```python
def is_palindrome(n):
    if len(n) < 1: #base
        return True
    else:
        if n[0] == n[-1]: #if the first and last character are same
            return is_palindrome(n[1:-1])
        else:
            return False
# Test cases
print(is_palindrome("madam"))     # True
print(is_palindrome("racecar"))   # True
print(is_palindrome("python"))    # False
```
