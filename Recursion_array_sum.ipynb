```python
def array_sum(arr, n):
    if n == 0:  #Base Case (Stopping Condition)
        return 0
    return arr[n-1] + array_sum(arr, n-1)

arr = [1, 2, 3, 4, 5]
print(array_sum(arr, len(arr)))  # Output: 15
```

ğŸ”¹ Base Case (Stopping Condition)


    if n == 0:
        return 0


Why is this needed?

    When there are 0 elements, the sum is 0

    Prevents infinite recursion

    Every recursive function must have a base case






ğŸ”¹ Recursive Case
    return arr[n-1] + array_sum(arr, n-1)

What is happening?

    Take the last element â†’ arr[n-1]

    Add it to the sum of the remaining (n-1) elements

    Call the same function again with a smaller problem

Step-by-Step Dry Run

  Array:
    arr = [1, 2, 3, 4, 5]

Call Stack Execution
    array_sum(arr, 5)
    = 5 + array_sum(arr, 4)

    array_sum(arr, 4)
    = 4 + array_sum(arr, 3)

    array_sum(arr, 3)
    = 3 + array_sum(arr, 2)

    array_sum(arr, 2)
    = 2 + array_sum(arr, 1)

    array_sum(arr, 1)
    = 1 + array_sum(arr, 0)

    array_sum(arr, 0)
    = 0   â† base case


Now returning back (stack unwinding)
    array_sum(arr, 1) = 1 + 0 = 1
    array_sum(arr, 2) = 2 + 1 = 3
    array_sum(arr, 3) = 3 + 3 = 6
    array_sum(arr, 4) = 4 + 6 = 10
    array_sum(arr, 5) = 5 + 10 = 15
