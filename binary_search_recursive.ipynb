üîπ *Problem*

Search an element in a **sorted array** using recursion.

To search an element in a sorted array, the most efficient approach is Binary Search.

üìå Why Binary Search?

Works only on sorted arrays

Faster than linear search

Reduces search space by half each time

‚è±Ô∏è Time & Space Complexity

Time Complexity: O(log n)

Space Complexity:

Iterative: O(1)

Recursive: O(log n) (call stack)

üî¢ Example

Array: [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
Target: 23
Output: Found at index 5

üß† Algorithm (Steps)

Set low = 0, high = n-1

Find mid = (low + high) // 2

If arr[mid] == target ‚Üí Found

If arr[mid] < target ‚Üí search right half

Else ‚Üí search left half

Repeat until found or range is empty

```python
def binary_search_recursive(arr, low, high, target):
    if low > high:
        return -1   # Element not found

    mid = (low + high) // 2 #Find middle index

    if arr[mid] == target: #Target found
        return mid
    elif arr[mid] > target:
        return binary_search_recursive(arr, low, mid - 1, target) #Target is smaller ‚Üí search left half
    else:
        return binary_search_recursive(arr, mid + 1, high, target) #Target is larger ‚Üí search right half


# Example
arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target = 23

result = binary_search_recursive(arr, 0, len(arr) - 1, target)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

üëâ Function receives:

arr ‚Üí sorted array

low ‚Üí start index

high ‚Üí end index

target ‚Üí number to find
