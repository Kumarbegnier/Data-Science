**Sum of Array using Recursion**

---



---



ðŸ”¹ Problem Statement

Find the sum of all elements in an array using recursion.

ðŸ”¹ Recursive Approach (Concept)

To sum an array recursively:

Base Case â†’ When the array has no elements, return 0

Recursive Case â†’
sum = first_element + sum(remaining_elements)

```python
def sum_array(arr, n):
    if n == 0: #Base Case â†’ When the array has no elements, return 0
      return 0
    return arr[0] + sum_array(arr[1:], n-1)

# test
arr = [1, 2, 3, 4, 5]
n = len(arr)

print("Sum of array:", sum_array(arr, n))
```
